### 图片的读取 ###


```python
import cv2
import matplotlib.pyplot as plt 
```

```python
jm = cv2.imread("JM.jpeg") # 默认读取彩色图片，BGR格式
print(jm)
```

```python
plt.imshow(jm) # matplotlib显示的图片和真实图片颜色不同，因为opencv读入的图片不是默认RGB图像，而是BGR图像，所以opencv读入的图片不要用别的模块展示
```

```python
cv2.imshow('johnmayer',jm)
key = cv2.waitKey(0)
if key == ord('q'):
    print("销毁图像...")
    cv2.destroyAllWindows()
    
```
